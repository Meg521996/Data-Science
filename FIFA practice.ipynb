```python
def get_best_squad_n(formation, nationality, measurement = 'Overall'):
    df_fifa_copy = df.copy()
    df_fifa_copy = df_fifa_copy[df_fifa_copy['Nationality'] == nationality]
    store = []
    
    for i in formation:
        store.append([
            df_fifa_copy.loc[[df_fifa_copy[df_fifa_copy['Position'].str.contains(i)][measurement].idxmax()]]['Position'].to_string(index = False),
            df_fifa_copy.loc[[df_fifa_copy[df_fifa_copy['Position'].str.contains(i)][measurement].idxmax()]]['Name'].to_string(index = False), 
            df_fifa_copy[df_fifa_copy['Position'].str.contains(i)][measurement].max(),
            df_fifa_copy.loc[[df_fifa_copy[df_fifa_copy['Position'].str.contains(i)][measurement].idxmax()]]['Age'].to_string(index = False),
            df_fifa_copy.loc[[df_fifa_copy[df_fifa_copy['Position'].str.contains(i)][measurement].idxmax()]]['Club'].to_string(index = False),
            df_fifa_copy.loc[[df_fifa_copy[df_fifa_copy['Position'].str.contains(i)][measurement].idxmax()]]['Value'].to_string(index = False),
            df_fifa_copy.loc[[df_fifa_copy[df_fifa_copy['Position'].str.contains(i)][measurement].idxmax()]]['Wage'].to_string(index = False)
        ])
        
        df_fifa_copy.drop(df_fifa_copy[df_fifa_copy['Position'].str.contains(i)][measurement].idxmax(), 
                         inplace = True)
    
    return np.mean([x[2] for x in store]).round(2), pd.DataFrame(np.array(store).reshape(11,7), 
                                                                 columns = ['Position', 'Player', measurement, 'Age', 'Club', 'Value', 'Wage']).to_string(index = False)
```

```python
def value_to_int(df_value):
    try:
        value = float(df_value[1:-1])
        suffix = df_value[-1:]

        if suffix == 'M':
            value = value * 1000000
        elif suffix == 'K':
            value = value * 1000
    except ValueError:
        value = 0
    return value

df['Value_float'] = df['Value'].apply(value_to_int)
```
